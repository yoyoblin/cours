### <span style="color:green"> Rappel : Une boucle `while` répète des instructions tant que la conditions choisie n'est pas atteinte... (Vous ne connaissez pas à l'avance le nombre de répétitions : c'est une boucle non bornée)

> - `while u <= 10 :` permet de répéter une boucle **tant que** la variable `u` n'a pas dépassé la valeur 10
> - **Il faut donner une valeur initiale** à `n` ET SURTOUT l'augmenter : **sinon vous créez une boucle infinie !**

```python
u = 3

while u <= 10 :
    print(u)
    u = u + 1
```
