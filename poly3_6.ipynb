### <span style="color:orange"> 1. Utiliser des listes en compréhension

> - **utilisation de l'expression d'une fonction :** par exemple, $f(x)=4\times x^{3}- 2\times x^{2}$ 
> - **pour des valeurs décimales** on peut utiliser le module **numpy** (sans le dire) et l'instruction `arange(début,fin,incrément)`

```python
from numpy import*
liste = [4*x**3 - 2*x**2 for x in arange(0,10.5,0.5)]

print(liste)
```

> - **Utilisation de fonction python :**

```python
def f(x) :
    return 4*x**3 - 2*x**2

liste = [(x,f(x)) for x in range(0,11)]

print(liste)
```

### <span style="color:orange"> 2. Approche de solution d'(in)équations $f(x)=c$ ou $f(x)≤c$ 

> - **utilisation de liste en compréhension** avec des possibilités de filtrer par des **tests conditionnels**
> - **Ici vous allez fixer une limite** pour approcher à 0.1 près une solution de l'inéquation $f(x)≤100$ 

```python
from numpy import*

def f(x) :
    return 4*x**3 - 2*x**2

liste = [(x,f(x)) for x in arange(2,3,0.1) if f(x) <= 100 ]

print(liste)
```
